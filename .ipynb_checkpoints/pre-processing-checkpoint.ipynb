```python
# Setting the Frequency
'''
'h' - hourly
'w' - weekly
'd' - daily
'b' - business days (without weekends)
'm' - monthly
'a' - annually
'''

df = df.asfreq('d')
df.head()
```

Since we generated new periods, some dates we do not have any values associated with them (NaN)

```python
# so we can use business days avoiding weekends
df = df.asfreq('b')
df.head()
```

No more NaN values for Weekends data since we removed weekends

### But we still have to check if there is anymore NaN values after we set the Frequency, eventhough original dataset did not have any missing dates when we inspect earlier

```python
df.isna().sum()
```

This confirms that after set the Frequency, there are 8 missing values

```python
# Filling those NaN values with one of the 3 methods
'''
1. Front Filling - Assigns the value of the previous period
2. Back Filling - Assigns the value of the next period
3. Assingning the same Value - Assign the average to all the missing values within the time series

* Front Filling copies the last known value, whereas backfilling uses tthe next known value

3rd Method is only make sense to perform if the data is heavily centered around the mean of the data in entire time series. 
Otherwise, it creates incorrect spikes or lows in the data which will affect for analysis 
'''

# Applying Font Filling Method for SPX column
df['spx'] = df['spx'].fillna(method='ffill')
df.isna().sum()
```

```python
# Applying Back Filling Method for FTSE column
df['ftse'] = df['ftse'].fillna(method='ffill')
df.isna().sum()
```

```python
# Applying Back Filling Method for FTSE column
df['dax'] = df['dax'].fillna(value=df['dax'].mean())
df.isna().sum()
```

```python
# Applying Font Filling Method for SPX column
df['nikkei'] = df['nikkei'].fillna(method='ffill')
df.isna().sum()
```

Experimented by applying 3 different filling methods to each column missing data

### Train and Test Split

Before feeding data into Machine Learning model, it is essential to split them into Train and Test Set

Also, important to shuffle the data BUT we cannot do it since it will mess up the Choronological order of the series

whole time series (starting_point-----------80%--------------cut_of_point----20%----end_point)
1. Training Set >> From the begining up to some cut off point (starting_point-------------cut_off_point)
2. Testing Set >> From the cut off point until the end (cut_off_point-----------end_point)

```python
cut_off_point = int(len(spx_df)*0.8)
spx_train = spx_df.iloc[:cut_off_point]
spx_test = spx_df.iloc[cut_off_point:]

spx_train.head(), spx_test.tail()
```
